### airkorea 인구 가중치 생성

```python
# airkorea data
data = pd.read_csv("/content/drive/MyDrive/날씨 빅콘/data/전처리데이터/대기오염/air_mice.csv")

# 인구 data
pop = pd.read_csv('/content/drive/MyDrive/날씨 빅콘/data/인구수/총인구수_air.csv')
```

```python
pop0 = pop.copy()
pop0.loc[244, "시군구"] = "세종"
pop0 = pop0.dropna(subset=["시군구"]).reset_index(drop=True)
```

```python
pd.set_option('display.max_rows', None)
```

```python
data['측정일']=data['측정일'].astype(str)
data['year']= data['측정일'].str[:4]
```

```python
con1 = data['시군구']!='강서로45다길'
data2 = data[con1]

con2 = data2['시군구']!='중구광복로'
data2 = data2[con2]

con3 = data2['시군구']!='강서로'
data2 = data2[con3]
```

```python
data2 = data2.reset_index(drop=True)
```

```python
data2.to_csv("/content/drive/MyDrive/날씨 빅콘/data/전처리데이터/대기오염/air_mice2.csv", index=False)
```

```python
df1 = data2[data2['year']=='2011']
df2 = data2[data2['year']=='2012']
df3 = data2[data2['year']=='2013']
df4 = data2[data2['year']=='2014']
df5 = data2[data2['year']=='2015']
df6 = data2[data2['year']=='2016']

df1 = df1[["지역","시군구"]]
df1.drop_duplicates(["지역","시군구"], keep = 'first', inplace=True)
df1= df1.reset_index(drop=True)

df2 = df2[["지역","시군구"]]
df2.drop_duplicates(["지역","시군구"], keep = 'first', inplace=True)
df2= df2.reset_index(drop=True)

df3 = df3[["지역","시군구"]]
df3.drop_duplicates(["지역","시군구"], keep = 'first', inplace=True)
df3= df3.reset_index(drop=True)

df4 = df4[["지역","시군구"]]
df4.drop_duplicates(["지역","시군구"], keep = 'first', inplace=True)
df4= df4.reset_index(drop=True)

df5 = df5[["지역","시군구"]]
df5.drop_duplicates(["지역","시군구"], keep = 'first', inplace=True)
df5= df5.reset_index(drop=True)

df6 = df6[["지역","시군구"]]
df6.drop_duplicates(["지역","시군구"], keep = 'first', inplace=True)
df6= df6.reset_index(drop=True)
```

```python
weight1 = pd.merge(df1, pop0[["지역","시군구","pop11"]], how = "left", on=["지역","시군구"])
weight2 = pd.merge(df2, pop0[["지역","시군구","pop12"]], how = "left", on=["지역","시군구"])
weight3 = pd.merge(df3, pop0[["지역","시군구","pop13"]], how = "left", on=["지역","시군구"])
weight4 = pd.merge(df4, pop0[["지역","시군구","pop14"]], how = "left", on=["지역","시군구"])
weight5 = pd.merge(df5, pop0[["지역","시군구","pop15"]], how = "left", on=["지역","시군구"])
weight6 = pd.merge(df6, pop0[["지역","시군구","pop16"]], how = "left", on=["지역","시군구"])
```

```python
weight = pd.merge(weight1, weight2, how="outer", on=["지역","시군구"])
weight = pd.merge(weight, weight3, how="outer", on=["지역","시군구"])
weight = pd.merge(weight, weight4, how="outer", on=["지역","시군구"])
weight = pd.merge(weight, weight5, how="outer", on=["지역","시군구"])
weight = pd.merge(weight, weight6, how="outer", on=["지역","시군구"])
```

```python
weight.head()
```

```python
pop1 = weight.groupby('지역').sum().reset_index()
```

```python
s1 = []
s2 = []
s3 = []
s4 = []
s5 = []
s6 = []

for i in range(163):
  area = weight['지역'][i]
  s1.append(pop1.loc[pop1['지역']==area, "pop11"].values[0])
  s2.append(pop1.loc[pop1['지역']==area, "pop12"].values[0])
  s3.append(pop1.loc[pop1['지역']==area, "pop13"].values[0])
  s4.append(pop1.loc[pop1['지역']==area, "pop14"].values[0])
  s5.append(pop1.loc[pop1['지역']==area, "pop15"].values[0])
  s6.append(pop1.loc[pop1['지역']==area, "pop16"].values[0])

weight['pop_sum11'] = s1
weight['pop_sum12'] = s2
weight['pop_sum13'] = s3
weight['pop_sum14'] = s4
weight['pop_sum15'] = s5
weight['pop_sum16'] = s6

weight['weight11'] = weight['pop11']/weight['pop_sum11']
weight['weight12'] = weight['pop12']/weight['pop_sum12']
weight['weight13'] = weight['pop13']/weight['pop_sum13']
weight['weight14'] = weight['pop14']/weight['pop_sum14']
weight['weight15'] = weight['pop15']/weight['pop_sum15']
weight['weight16'] = weight['pop16']/weight['pop_sum16']
```

```python
weight = weight[['지역','시군구','weight11', 'weight12', 'weight13', 'weight14', 'weight15','weight16']]
```
