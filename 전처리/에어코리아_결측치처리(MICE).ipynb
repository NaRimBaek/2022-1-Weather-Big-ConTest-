## 강원

```python
gangwon = air[(air['지역']=="강원")]
gangwon_train = gangwon[(gangwon['측정일']>=20110101) & (gangwon['측정일']<=20151231)]
gangwon_test = gangwon[gangwon['측정일']>=20160101]
gangwon_train2 = gangwon_train.iloc[:, 3:]
gangwon_test2 = gangwon_test.iloc[:, 3:]
```

```python
gangwon
```

```python
gangwon_train.columns
```

```python
gangwon_train
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# random_state값은 원하시는 숫자 아무거나 넣으시면 됩니다.
imputer_mice = IterativeImputer(random_state=24)
gangwon_train_mice = imputer_mice.fit_transform(gangwon_train2)
gangwon_test_mice = imputer_mice.transform(gangwon_test2)
```

```python
df1 = pd.DataFrame(gangwon_train_mice, columns = ['SO2', 'CO', 'O3', 'NO2', 'PM10'])
df2 = pd.DataFrame(gangwon_test_mice, columns = ['SO2', 'CO', 'O3', 'NO2', 'PM10'])
df3 = pd.concat([df1, df2], ignore_index=True)
df3
```

```python
df = pd.concat([df3, gangwon.iloc[:, 0:3]], axis=1 )
df = df[['지역', '측정일', '시군구', 'SO2', 'CO', 'O3', 'NO2', 'PM10']]
df
```

```python
df.to_csv('/content/drive/MyDrive/날씨 빅콘/data/전처리데이터/대기오염/대기오염mice/강원_대기오염_mice.csv', index=False)
```
