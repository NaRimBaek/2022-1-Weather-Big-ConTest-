```python
#가중치 데이터 붙이기
dat2=pd.merge(dat, weight, how='left', on=['area','region'])

# 연도 칼럼 만들기
dat2['year']=dat2.date.str[:4]

# 연도별 데이터로 나누기
dat2012=dat2[dat2['year']=='2012']
dat2013=dat2[dat2['year']=='2013']
dat2014=dat2[dat2['year']=='2014']
dat2015=dat2[dat2['year']=='2015']
dat2016=dat2[dat2['year']=='2016']

# 해당 연도 가중치 컬럼만 남기기
dat2012_2=dat2012.drop(['weight_13','weight_14','weight_15','weight_16'],axis=1).copy()
dat2013_2=dat2013.drop(['weight_12','weight_14','weight_15','weight_16'],axis=1).copy()
dat2014_2=dat2014.drop(['weight_12','weight_13','weight_15','weight_16'],axis=1).copy()
dat2015_2=dat2015.drop(['weight_12','weight_13','weight_14','weight_16'],axis=1).copy()
dat2016_2=dat2016.drop(['weight_12','weight_13','weight_14','weight_15'],axis=1).copy()

# 가중치 곱한 연도별 데이터 만들기
for i in range(1,11):
  dat2012_2.iloc[:,i]=dat2012_2.iloc[:,i]*dat2012_2.iloc[:,14]

for i in range(1,11):
  dat2013_2.iloc[:,i]=dat2013_2.iloc[:,i]*dat2013_2.iloc[:,14]

for i in range(1,11):
  dat2014_2.iloc[:,i]=dat2014_2.iloc[:,i]*dat2014_2.iloc[:,14]

for i in range(1,11):
  dat2015_2.iloc[:,i]=dat2015_2.iloc[:,i]*dat2015_2.iloc[:,14]

for i in range(1,11):
  dat2016_2.iloc[:,i]=dat2016_2.iloc[:,i]*dat2016_2.iloc[:,14]

# 날짜별 groupby sum

inc2012=dat2012_2.groupby(dat2012_2['date']).sum().reset_index().drop(['weight_12'],axis=1)
inc2013=dat2013_2.groupby(dat2013_2['date']).sum().reset_index().drop(['weight_13'],axis=1)
inc2014=dat2014_2.groupby(dat2014_2['date']).sum().reset_index().drop(['weight_14'],axis=1)
inc2015=dat2015_2.groupby(dat2015_2['date']).sum().reset_index().drop(['weight_15'],axis=1)
inc2016=dat2016_2.groupby(dat2016_2['date']).sum().reset_index().drop(['weight_16'],axis=1)

# 시군구 하나로 합치기
inc=pd.concat([inc2012,inc2013,inc2014,inc2015,inc2016])
inc.head()
```
