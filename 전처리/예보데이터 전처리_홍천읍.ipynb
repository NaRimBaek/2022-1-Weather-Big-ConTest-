```python
def make_date(data, year, month):
  num = data.shape[0]
  df = data.reset_index()
  list1 = list(pd.to_numeric(df['day']))
  df['date'] = df['day']

  for i in range(num):
    df['date'][i] = dt.datetime(year, month, list1[i])
  
  df['date'] = pd.to_datetime(df['date'], format = '%Y%m%d')
  df = df.iloc[:,1:]
  return df
```
