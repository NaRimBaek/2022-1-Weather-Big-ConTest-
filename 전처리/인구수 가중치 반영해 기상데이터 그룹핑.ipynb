## 예보데이터를 지역별로 MICE 처리 후 저장하기

```python
######################### 데이터 다 다운받은 다음 다시 돌리기 #########################
###############################################################################
region_list = os.listdir(path + '/전처리데이터/예보데이터') 


for region in region_list: 

  # 파일 경로 
  region_path = path + '/전처리데이터/예보데이터/' + region
  print('지역명: ', region) 

  file_list2 = os.listdir(region_path) 

  dong_list = list(set( [i.split('_')[1] for i in file_list2] ))
  print('읍면동: ', dong_list)
  print('읍면동 개수: ', len(dong_list))
  
  print()
```
