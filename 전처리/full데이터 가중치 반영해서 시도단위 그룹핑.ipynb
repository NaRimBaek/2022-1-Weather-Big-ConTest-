```python
import os
import numpy as np
import pandas as pd
import datetime as dt
```

```python
# 관측 full data
data = pd.read_csv('/content/drive/MyDrive/날씨 빅콘/data/전처리데이터/full_전처리완료3.csv')
sejong = pd.read_csv('/content/drive/MyDrive/날씨 빅콘/data/전처리데이터/sejong_full.csv')
# 가중치 data
weight = pd.read_csv('/content/drive/MyDrive/날씨 빅콘/data/인구수/weight_full.csv')
```

```python
# 가중치 곱하기 전 제외할 변수 

var = ['stn_id', 'hr6_max_rn_dt', 'mix_ws','mix_wd','max_ws_wd','max_ins_ws_wd', 'hr24_sum_rws', 'max_wd', 'max_ws_min', 'max_ws_hour', 'max_ins_ws_hour', 'max_ins_ws_min'   ]
```

```python
# 그룹핑 기준 변수
# where, tma 
# pd.set_option('display.max_rows',None)
```

```python
data2 = data.drop(var, axis=1).reset_index(drop=True)
```

```python
weight.head()
```

```python
dat2011 = data2[data2['year']==2011]
dat2011_2 = pd.merge(dat2011, weight[['where','stn', 'weight11']], how='left', on=["where",'stn'])

dat2012 = data2[data2['year']==2012]
dat2012_2 = pd.merge(dat2012, weight[['where','stn', 'weight12']], how='left', on=["where",'stn'])

dat2013 = data2[data2['year']==2013]
dat2013_2 = pd.merge(dat2013, weight[['where','stn', 'weight13']], how='left', on=["where",'stn'])

dat2014 = data2[data2['year']==2014]
dat2014_2 = pd.merge(dat2014, weight[['where','stn', 'weight14']], how='left', on=["where",'stn'])

dat2015 = data2[data2['year']==2015]
dat2015_2 = pd.merge(dat2015, weight[['where','stn', 'weight15']], how='left', on=["where",'stn'])

dat2016 = data2[data2['year']==2016]
dat2016_2 = pd.merge(dat2016, weight[['where','stn', 'weight16']], how='left', on=["where",'stn'])
```

```python
for i in range(1,28):
  dat2011_2.iloc[:,i] = dat2011_2.iloc[:,i] * dat2011_2.iloc[:,33]
dat2011_3 = dat2011_2.groupby(['where','tma']).sum().reset_index().drop(['year', 'month', 'days', 'weight11'], axis=1)

for i in range(1,28):
  dat2012_2.iloc[:,i] = dat2012_2.iloc[:,i] * dat2012_2.iloc[:,33]
dat2012_3 = dat2012_2.groupby(['where','tma']).sum().reset_index().drop(['year', 'month', 'days', 'weight12'], axis=1)

for i in range(1,28):
  dat2013_2.iloc[:,i] = dat2013_2.iloc[:,i] * dat2013_2.iloc[:,33]
dat2013_3 = dat2013_2.groupby(['where','tma']).sum().reset_index().drop(['year', 'month', 'days', 'weight13'], axis=1)

for i in range(1,28):
  dat2014_2.iloc[:,i] = dat2014_2.iloc[:,i] * dat2014_2.iloc[:,33]
dat2014_3 = dat2014_2.groupby(['where','tma']).sum().reset_index().drop(['year', 'month', 'days', 'weight14'], axis=1)

for i in range(1,28):
  dat2015_2.iloc[:,i] = dat2015_2.iloc[:,i] * dat2015_2.iloc[:,33]
dat2015_3 = dat2015_2.groupby(['where','tma']).sum().reset_index().drop(['year', 'month', 'days', 'weight15'], axis=1)

for i in range(1,28):
  dat2016_2.iloc[:,i] = dat2016_2.iloc[:,i] * dat2016_2.iloc[:,33]
dat2016_3 = dat2016_2.groupby(['where','tma']).sum().reset_index().drop(['year', 'month', 'days', 'weight16'], axis=1)
```

```python
df = pd.concat([dat2011_3, dat2012_3], ignore_index=True)
df = pd.concat([df, dat2013_3], ignore_index=True)
df = pd.concat([df, dat2014_3], ignore_index=True)
df = pd.concat([df, dat2015_3], ignore_index=True)
df = pd.concat([df, dat2016_3], ignore_index=True)
```
