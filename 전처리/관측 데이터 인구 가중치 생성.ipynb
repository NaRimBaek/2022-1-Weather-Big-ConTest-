```python
import os
import numpy as np
import pandas as pd
import datetime as dt
```

```python
# 관측 full data
data = pd.read_csv('/content/drive/MyDrive/날씨 빅콘/data/전처리데이터/full_전처리완료3.csv')

# 인구수 data
pop = pd.read_csv('/content/drive/MyDrive/날씨 빅콘/data/인구수/총인구수_full.csv')
```

```python
df1 = data[data['year']==2011]
df2 = data[data['year']==2014]

df1 = df1[["where","stn"]]
df1.drop_duplicates(["where","stn"], keep = 'first', inplace=True)
df1= df1.reset_index(drop=True)

df2 = df2[["where","stn"]]
df2.drop_duplicates(["where","stn"], keep = 'first', inplace=True)
df2= df2.reset_index(drop=True)
```

```python
weight1 = pd.merge(df1, pop[["지역","시군구","pop11","pop12","pop13"]], how = "left", left_on=["where","stn"], right_on=["지역","시군구"])
var1 = weight1.loc[weight1['pop11'].isnull(), 'stn'].unique()

for name in var1:
  var2 = name[:2]
  con = weight1['stn']== name
  weight1.loc[con, "pop11"] = pop.loc[pop['시군구']==var2, "pop11"].values[0]
  weight1.loc[con, "pop12"] = pop.loc[pop['시군구']==var2, "pop12"].values[0]
  weight1.loc[con, "pop13"] = pop.loc[pop['시군구']==var2, "pop13"].values[0]



weight2 = pd.merge(df2, pop[["지역","시군구","pop14","pop15","pop16"]], how = "left", left_on=["where","stn"], right_on=["지역","시군구"])
var3 = weight2.loc[weight2['pop14'].isnull(), 'stn'].unique()

for name in var3:
  var4 = name[:2]
  con = weight2['stn']== name
  weight2.loc[con, "pop14"] = pop.loc[pop['시군구']==var4, "pop14"].values[0]
  weight2.loc[con, "pop15"] = pop.loc[pop['시군구']==var4, "pop15"].values[0]
  weight2.loc[con, "pop16"] = pop.loc[pop['시군구']==var4, "pop16"].values[0]
  
```

```python
weight1.isnull().sum()
```

```python
weight2.isnull().sum()
```

```python
weight1 = weight1.drop(["지역","시군구"], axis=1)
weight2 = weight2.drop(["지역","시군구"], axis=1)
weight = pd.merge(weight1, weight2, how="outer", on=["where","stn"])
```

```python
pop1 = weight.groupby('where').sum().reset_index()
```

```python
pop1
```

```python
weight.shape
```

```python
s1 = []
s2 = []
s3 = []
s4 = []
s5 = []
s6 = []

for i in range(83):
  area = weight['where'][i]
  s1.append(pop1.loc[pop1['where']==area, "pop11"].values[0])
  s2.append(pop1.loc[pop1['where']==area, "pop12"].values[0])
  s3.append(pop1.loc[pop1['where']==area, "pop13"].values[0])
  s4.append(pop1.loc[pop1['where']==area, "pop14"].values[0])
  s5.append(pop1.loc[pop1['where']==area, "pop15"].values[0])
  s6.append(pop1.loc[pop1['where']==area, "pop16"].values[0])

weight['pop_sum11'] = s1
weight['pop_sum12'] = s2
weight['pop_sum13'] = s3
weight['pop_sum14'] = s4
weight['pop_sum15'] = s5
weight['pop_sum16'] = s6

weight['weight11'] = weight['pop11']/weight['pop_sum11']
weight['weight12'] = weight['pop12']/weight['pop_sum12']
weight['weight13'] = weight['pop13']/weight['pop_sum13']
weight['weight14'] = weight['pop14']/weight['pop_sum14']
weight['weight15'] = weight['pop15']/weight['pop_sum15']
weight['weight16'] = weight['pop16']/weight['pop_sum16']
```

```python
weight_full = weight[['where','stn','weight11', 'weight12', 'weight13', 'weight14', 'weight15','weight16']]
```
