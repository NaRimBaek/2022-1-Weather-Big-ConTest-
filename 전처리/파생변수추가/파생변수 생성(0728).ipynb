급성심장정지조사

```python
dat=pd.read_csv('/content/drive/MyDrive/날씨 빅콘/data/파생변수/급성심정지/12-19 지역,성별별 급성심정지 전체_발생건수_20220728162147.csv',encoding='cp949')
```

```python
dat.rename(columns={'시도별(1)':'area'},inplace=True)
dat.rename(columns={'성별 및 연령별(2)':'sex'},inplace=True)
dat.rename(columns={'발생건수 (건)':'심정지'},inplace=True)
dat.rename(columns={'시점':'year'},inplace=True)
dat.head()
```

```python
condition = (dat.area!='전국') & ((dat.sex=='여자') |(dat.sex=='남자') )
dat = dat[condition][['year','area','sex','심정지']]
dat
```

```python
dat.loc[dat['sex'] == '남자', 'sex'] = 1
dat.loc[dat['sex'] == '여자', 'sex'] = 2
dat
```
