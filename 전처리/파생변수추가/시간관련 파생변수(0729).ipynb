```python
train['day']=train['yyyymmdd'].str[8:10]
test['day']=test['yyyymmdd'].str[8:10]
```

```python
train['day1']=0

for i in range(train.shape[0]):
  if train['day'][i]=='01':
    train['day1'][i]=1
  else:
    train['day1'][i]=0
```

```python
test['day1']=0

for i in range(test.shape[0]):
  if test['day'][i]=='01':
    test['day1'][i]=1
  else:
    test['day1'][i]=0
```
