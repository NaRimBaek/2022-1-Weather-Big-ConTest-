```python
import os
import pandas as pd
import numpy as np
```

## 코드

```python
smoke = pd.read_csv("/content/drive/MyDrive/날씨 빅콘/data/건강데이터/지역별_성별_흡연.csv", encoding = 'cp949' )
```

```python
smoke2 = smoke.loc[((smoke['성별(1)'] == '남자') | (smoke['성별(1)'] == '여자')) & (smoke['시도별(1)']!="계"),:]
smoke_sex = smoke2.iloc[:,0:2]
smoke_sex.columns = ['area', 'sex']

smoke2012 = smoke2.loc[:,smoke2.columns.str.startswith("2012.")]
smoke2013 = smoke2.loc[:,smoke2.columns.str.startswith("2013.")]
smoke2014 = smoke2.loc[:,smoke2.columns.str.startswith("2014.")]
smoke2015 = smoke2.loc[:,smoke2.columns.str.startswith("2015.")]
smoke2016 = smoke2.loc[:,smoke2.columns.str.startswith("2016.")]

smoke2012_1 = smoke2012[['2012.4', '2012.7','2012.8']]
smoke2013_1 = smoke2013[['2013.4', '2013.7','2013.8']]
smoke2014_1 = smoke2014[['2014.4', '2014.7','2014.8']]
smoke2015_1 = smoke2015[['2015.4', '2015.7','2015.8']]
smoke2016_1 = smoke2016[['2016.4', '2016.7','2016.8']]

smoke2012_1.columns = ['smoking','day_mean_smoke1','day_mean_smoke2'] # 현재 흡연자, 하루평균 한갑 이상 두 갑 미만, 두갑 이상
smoke2013_1.columns = ['smoking','day_mean_smoke1','day_mean_smoke2']
smoke2014_1.columns = ['smoking','day_mean_smoke1','day_mean_smoke2']
smoke2015_1.columns = ['smoking','day_mean_smoke1','day_mean_smoke2']
smoke2016_1.columns = ['smoking','day_mean_smoke1','day_mean_smoke2']

smoke2012_1['year'] = 2012
smoke2013_1['year'] = 2013
smoke2014_1['year'] = 2014 
smoke2015_1['year'] = 2015 
smoke2016_1['year'] = 2016

smoke2012_2 = pd.concat([smoke_sex,smoke2012_1], axis=1)
smoke2013_2 = pd.concat([smoke_sex,smoke2013_1], axis=1)
smoke2014_2 = pd.concat([smoke_sex,smoke2014_1], axis=1)
smoke2015_2 = pd.concat([smoke_sex,smoke2015_1], axis=1)
smoke2016_2 = pd.concat([smoke_sex,smoke2016_1], axis=1)

df = pd.concat([smoke2012_2, smoke2013_2])
df1 = pd.concat([df, smoke2014_2])
df2 = pd.concat([df1, smoke2015_2])
df3 = pd.concat([df2, smoke2016_2])

df3_2 = df3.copy()
df3_2.iloc[15,4] = 0  # 결측치 채우기

df3_2['smoking'] = df3_2['smoking'].astype(int)
df3_2['day_mean_smoke1'] = df3_2['day_mean_smoke1'].astype(int)
df3_2['day_mean_smoke2'] = df3_2['day_mean_smoke2'].astype(int)
df3_2['day_mean_smoke_over1'] = df3_2['day_mean_smoke1'] + df3_2['day_mean_smoke2']
df3_2 = df3_2.drop(['day_mean_smoke1', 'day_mean_smoke2'], axis=1)
df3_2 = df3_2.reset_index(drop=True)
```

```python
df3_2.to_csv('/content/drive/MyDrive/날씨 빅콘/data/파생변수/완성된 파생변수들/흡연.csv', index=False)
```
