```python
disease2 = disease.loc[((disease['성별(1)'] == '남자') | (disease['성별(1)'] == '여자')) & (disease['시도별(1)']!="계"),:]

disease_sex = disease2.iloc[:,0:2]
disease2012 = disease2.loc[:,disease2.columns.str.startswith("2012.")]
disease2013 = disease2.loc[:,disease2.columns.str.startswith("2013.")]
disease2014 = disease2.loc[:,disease2.columns.str.startswith("2014.")]
disease2015 = disease2.loc[:,disease2.columns.str.startswith("2015.")]
disease2016 = disease2.loc[:,disease2.columns.str.startswith("2016.")]
```

```python
disease_sex.columns = ['area', 'sex']
disease_sex.head(10)
```

```python
disease2016.columns
```

```python
disease2012['year'] = 2012
disease2013['year'] = 2013
disease2014['year'] = 2014
disease2015['year'] = 2015
disease2016['year'] = 2016


disease2012_1 = disease2012[['year','2012.1', '2012.2', '2012.3', '2012.5','2012.9', '2012.10', '2012.11','2012.13','2012.17', '2012.18', '2012.19']]
disease2013_1 = disease2013[['year','2013.1', '2013.2', '2013.3', '2013.5','2013.9', '2013.10', '2013.11','2013.13','2013.17', '2013.18', '2013.19']]
disease2014_1 = disease2014[['year','2014.1', '2014.2', '2014.3', '2014.5','2014.9', '2014.10', '2014.11','2014.13','2014.17', '2014.18', '2014.19']]
disease2015_1 = disease2015[['year','2015.1', '2015.2', '2015.3', '2015.5','2015.9', '2015.10', '2015.11','2015.13','2015.17', '2015.18', '2015.19']]
disease2016_1 = disease2016[['year','2016.1', '2016.2', '2016.3', '2016.5','2016.9', '2016.10', '2016.11','2016.13','2016.17', '2016.18', '2016.19']]
```

```python
col = ['year','뇌졸중_진단','심장병_진단','고혈압_진단','이상지질혈증_진단',"뇌졸중_약물치료",'심장병_약물치료','고혈압_약물치료','이상지질혈증_약물치료','뇌졸중_가족력','심장병_가족력','고혈압_가족력']

disease2012_1.columns = col
disease2013_1.columns = col
disease2014_1.columns = col
disease2015_1.columns = col
disease2016_1.columns = col
```

```python
disease2012_1.head()
```

```python
df1 = pd.concat([disease_sex,disease2012_1], axis=1)
df2 = pd.concat([disease_sex,disease2013_1], axis=1)
df3 = pd.concat([disease_sex,disease2014_1], axis=1)
df4 = pd.concat([disease_sex,disease2015_1], axis=1)
df5 = pd.concat([disease_sex,disease2016_1], axis=1)

df_list = [df1, df2, df3, df4, df5]
final_data = pd.concat(df_list,axis=0,ignore_index=True)
final_data.head()
```

```python
final_data.loc[final_data['sex']=="남자", 'sex'] = 1
final_data.loc[final_data['sex']=="여자", 'sex'] = 2
final_data.head()
```
