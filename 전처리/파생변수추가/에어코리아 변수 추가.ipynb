```python
import os
import numpy as np
import pandas as pd
import datetime as dt
```

```python
# 관측 full data
#fulldata = pd.read_csv('/content/drive/MyDrive/날씨 빅콘/data/전처리데이터/full_가중치완료.csv')

# 인구수 data
airdata = pd.read_csv('/content/drive/MyDrive/날씨 빅콘/data/전처리데이터/air_가중치완료.csv')
```

```python
airdata['측정일'] = airdata['측정일'].astype(str)
airdata['측정일'] = airdata['측정일'].str[:4] + '-' + airdata['측정일'].str[4:6] + '-' + airdata['측정일'].str[6:]
airdata['측정일'] = pd.to_datetime(airdata['측정일'] )
```

```python
# lag1  (1/1 -> 1/2) 
# lag2  (1/1 -> 1/3)
# lag3  (1/1 -> 1/4)
# na값은 뒤에서 채워오기
```

```python
lag_columns = ['SO2', 'CO', 'O3', 'NO2', 'PM10']
lag1_columns = ['S02_lag1','CO_lag1','O3_lag1','NO2_lag1','PM10_lag1']
lag2_columns = ['S02_lag2','CO_lag2','O3_lag2','NO2_lag2','PM10_lag2']
lag3_columns = ['S02_lag3','CO_lag3','O3_lag3','NO2_lag3','PM10_lag3']
```

```python
airdata[lag1_columns] = airdata.groupby(["지역"])[['SO2', 'CO', 'O3', 'NO2', 'PM10']].shift(1)
airdata[lag2_columns] = airdata.groupby(["지역"])[['SO2', 'CO', 'O3', 'NO2', 'PM10']].shift(2)
airdata[lag3_columns] = airdata.groupby(["지역"])[['SO2', 'CO', 'O3', 'NO2', 'PM10']].shift(3)
```

```python
for col in lag1_columns:
 airdata[col] = airdata[col].fillna(airdata[col][1]) 

for col in lag2_columns:
 airdata[col] = airdata[col].fillna(airdata[col][2]) 

for col in lag3_columns:
 airdata[col] = airdata[col].fillna(airdata[col][3]) 
```
