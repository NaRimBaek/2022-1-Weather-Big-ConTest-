```python
population['2012년_계_연령구간인구수'] = population['2012년_계_연령구간인구수'].str.replace(',', '').astype('int64')
population['2013년_계_연령구간인구수'] = population['2013년_계_연령구간인구수'].str.replace(',', '').astype('int64') 
population['2014년_계_연령구간인구수'] = population['2014년_계_연령구간인구수'].str.replace(',', '').astype('int64') 
population['2015년_계_연령구간인구수'] = population['2015년_계_연령구간인구수'].str.replace(',', '').astype('int64') 
population['2016년_계_연령구간인구수'] = population['2016년_계_연령구간인구수'].str.replace(',', '').astype('int64') 
```

```python
population.info()
```

```python
population = population.rename(columns={'행정구역':'area'})
```

```python
pop_2012 = population[['area','2012년_계_연령구간인구수']]
pop_2013 = population[['area','2013년_계_연령구간인구수']]
pop_2014 = population[['area','2014년_계_연령구간인구수']]
pop_2015 = population[['area','2015년_계_연령구간인구수']]
pop_2016 = population[['area','2016년_계_연령구간인구수']]
```

```python
pop_2012['year']=2012
pop_2013['year']=2013
pop_2014['year']=2014
pop_2015['year']=2015
pop_2016['year']=2016
```

```python
pop_2012 = pop_2012.rename(columns={'2012년_계_연령구간인구수':'pop_over50'})
pop_2013 = pop_2013.rename(columns={'2013년_계_연령구간인구수':'pop_over50'})
pop_2014 = pop_2014.rename(columns={'2014년_계_연령구간인구수':'pop_over50'})
pop_2015 = pop_2015.rename(columns={'2015년_계_연령구간인구수':'pop_over50'})
pop_2016 = pop_2016.rename(columns={'2016년_계_연령구간인구수':'pop_over50'})
```

```python
pop = pd.concat([pop_2012,pop_2013,pop_2014,pop_2015,pop_2016])
```

```python
train = pd.read_csv("/content/drive/MyDrive/날씨 빅콘/data/전처리데이터/train.csv")
test = pd.read_csv("/content/drive/MyDrive/날씨 빅콘/data/전처리데이터/test.csv")
```

```python
df = pd.concat([train, test], axis=0, ignore_index=True)
df['yyyymmdd'] = pd.to_datetime(df['yyyymmdd'], format='%Y-%m-%d')
```

```python
df['year'] = df['yyyymmdd'].dt.year 
```

```python
df2 = pd.merge(df, pop, how="left",on=['area', 'year'])
df2
```

```python
df2 = df2.drop(['year'],axis=1)
```

```python
train2 = df2.loc[df2['yyyymmdd'].dt.year!=2016,:]
test2 = df2.loc[df2['yyyymmdd'].dt.year==2016,:]
```

```python
test2 = test2.reset_index().drop('index', axis=1)
test2
```

```python
test2.drop('frequency', axis=1, inplace=True)
test2
```

```python
train2.to_csv('/content/drive/MyDrive/날씨 빅콘/data/전처리데이터/train_pop.csv', index=False)
test2.to_csv("/content/drive/MyDrive/날씨 빅콘/data/전처리데이터/test_pop.csv", index=False)
```
