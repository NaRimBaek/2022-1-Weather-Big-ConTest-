```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
dj_mice = pd.read_csv('/content/drive/MyDrive/날씨 빅콘/data/전처리데이터/대기오염/대기오염mice/대전_대기오염_mice.csv')
cn_mice = pd.read_csv('/content/drive/MyDrive/날씨 빅콘/data/전처리데이터/대기오염/대기오염mice/충남_대기오염_mice.csv')
cb_mice = pd.read_csv('/content/drive/MyDrive/날씨 빅콘/data/전처리데이터/대기오염/대기오염mice/충북_대기오염_mice.csv')
```

```python
dj_mice2 = dj_mice.loc[dj_mice['측정일']>=20120101,:]
dj_mice2
```

```python
cn_mice['시군구'].unique()
```

```python
cn_mice2 = cn_mice.loc[((cn_mice['시군구']=='공주시')|(cn_mice['시군구']=='아산시')|(cn_mice['시군구']=='천안시')) & (cn_mice['측정일']>=20120101),:]
cn_mice2.head()
```

```python
cb_mice2 = cb_mice.loc[(cb_mice['시군구'] == '청주시') & (cb_mice['측정일']>=20120101),:]
cb_mice2.head()
```

```python
df_list = [dj_mice2, cn_mice2, cb_mice2]

df1 = pd.concat(df_list, axis=0, ignore_index=True)
df1
```

```python
df2 = df1.groupby('측정일',as_index=False).mean()[['측정일', 'SO2', 'CO', 'O3', 'NO2', 'PM10']]
df2['지역'] = '세종'
df2 = df2[['지역', '측정일', 'SO2', 'CO', 'O3', 'NO2', 'PM10']]
df2
```

```python
df2.to_csv('/content/drive/MyDrive/날씨 빅콘/data/전처리데이터/대기오염/대기오염mice/세종_대기오염_mice.csv', index=False)
```
