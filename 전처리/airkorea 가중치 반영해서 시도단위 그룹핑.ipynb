```python
dat2011 = data[data['year']==2011]
dat2011_2 = pd.merge(dat2011, weight[['지역','시군구', 'weight11']], how='left', on=['지역','시군구'])

dat2012 = data[data['year']==2012]
dat2012_2 = pd.merge(dat2012, weight[['지역','시군구', 'weight12']], how='left', on=['지역','시군구'])

dat2013 = data[data['year']==2013]
dat2013_2 = pd.merge(dat2013, weight[['지역','시군구', 'weight13']], how='left', on=['지역','시군구'])

dat2014 = data[data['year']==2014]
dat2014_2 = pd.merge(dat2014, weight[['지역','시군구', 'weight14']], how='left', on=['지역','시군구'])

dat2015 = data[data['year']==2015]
dat2015_2 = pd.merge(dat2015, weight[['지역','시군구', 'weight15']], how='left', on=['지역','시군구'])

dat2016 = data[data['year']==2016]
dat2016_2 = pd.merge(dat2016, weight[['지역','시군구', 'weight16']], how='left', on=['지역','시군구'])
```

```python
for i in range(3,8):
  dat2011_2.iloc[:,i] = dat2011_2.iloc[:,i] * dat2011_2.iloc[:,9]
dat2011_3 = dat2011_2.groupby(['지역','측정일']).sum().reset_index().drop(['year', 'weight11'], axis=1)

for i in range(3,8):
  dat2012_2.iloc[:,i] = dat2012_2.iloc[:,i] * dat2012_2.iloc[:,9]
dat2012_3 = dat2012_2.groupby(['지역','측정일']).sum().reset_index().drop(['year',  'weight12'], axis=1)

for i in range(3,8):
  dat2013_2.iloc[:,i] = dat2013_2.iloc[:,i] * dat2013_2.iloc[:,9]
dat2013_3 = dat2013_2.groupby(['지역','측정일']).sum().reset_index().drop(['year', 'weight13'], axis=1)

for i in range(3,8):
  dat2014_2.iloc[:,i] = dat2014_2.iloc[:,i] * dat2014_2.iloc[:,9]
dat2014_3 = dat2014_2.groupby(['지역','측정일']).sum().reset_index().drop(['year', 'weight14'], axis=1)

for i in range(3,8):
  dat2015_2.iloc[:,i] = dat2015_2.iloc[:,i] * dat2015_2.iloc[:,9]
dat2015_3 = dat2015_2.groupby(['지역','측정일']).sum().reset_index().drop(['year', 'weight15'], axis=1)

for i in range(3,8):
  dat2016_2.iloc[:,i] = dat2016_2.iloc[:,i] * dat2016_2.iloc[:,9]
dat2016_3 = dat2016_2.groupby(['지역','측정일']).sum().reset_index().drop(['year', 'weight16'], axis=1)
```

```python
df = pd.concat([dat2011_3, dat2012_3], ignore_index=True)
df = pd.concat([df, dat2013_3], ignore_index=True)
df = pd.concat([df, dat2014_3], ignore_index=True)
df = pd.concat([df, dat2015_3], ignore_index=True)
df = pd.concat([df, dat2016_3], ignore_index=True)
```
